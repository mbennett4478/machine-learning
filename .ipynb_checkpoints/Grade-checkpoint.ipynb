```python
count = 0

def zeroCount(dataFrame, columnName):
    noOfZeros = 0
    for i in range(dataFrame.shape[0]):
        if dataFrame[columnName][i] == 0:
            noOfZeros = noOfZeros + 1
    return noOfZeros

count = zeroCount(df, ' n_unique_tokens')

print("Number of 0s in the Target Column: %d" % count)
```

```python
df.dropna()
```

```python
df = df[~df[' n_unique_tokens'].isin([0, 1])]
```

```python
df[" n_unique_tokens"] = df[" n_unique_tokens"].fillna(0)
```

```python
df[" n_unique_tokens"]
```

```python
newCount = 0

def zeroCount(dataFrame, columnName):
    noOfZeros = 0
    for i in range(len(dataFrame.index.get_values())):
        if dataFrame[columnName][dataFrame.index.get_values()[i]] == 0:
            noOfZeros = noOfZeros + 1
    return noOfZeros

newCount = zeroCount(df, ' n_unique_tokens')

print("Number of 0s in the Target Column: %d" % newCount)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = (9, 9)

# Bar plot of grades
plt.bar(df[' shares'].value_counts().index, 
        df[' shares'].value_counts().values,
         fill = 'navy', edgecolor = 'k', width = 2)
plt.xlabel('Shares'); 
plt.ylabel('Count'); 
plt.title('Distribution of Unique Tokens');
plt.xticks(list(range(5, 20)));
```

```python
from scipy.stats import percentileofscore


# Calculate percentile for grades
df['percentile'] = df[' shares'].apply(lambda x: percentileofscore(df[' shares'], x))

# Plot percentiles for grades
plt.figure(figsize = (8, 6))

plt.plot(df[' shares'], df['percentile'], 'o')

plt.xticks(range(0, 20, 2), range(0, 20, 2))

plt.xlabel('Share'); 
plt.ylabel('Percentile'); 
plt.title('Share Percentiles');
```
